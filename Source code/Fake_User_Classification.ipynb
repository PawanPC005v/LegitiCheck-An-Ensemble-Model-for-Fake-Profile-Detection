```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler, LabelEncoder
from xgboost import XGBClassifier
import seaborn as sns

# Step 1: Data Acquisition and Preprocessing
def read_datasets(real_users_path="/content/users.csv", fake_users_path="/content/fusers.csv"):
    """
    Reads datasets for real and fake users, combines them, and creates labels.
    """
    try:
        # Load datasets
        real_users = pd.read_csv(real_users_path)
        fake_users = pd.read_csv(fake_users_path)

        # Combine datasets
        x = pd.concat([real_users, fake_users], ignore_index=True)

        # Create labels (0 for fake, 1 for real)
        y = [0] * len(fake_users) + [1] * len(real_users)

        print(f"Datasets loaded successfully: Real users ({len(real_users)}), Fake users ({len(fake_users)})")
        return x, y

    except FileNotFoundError as e:
        print(f"Error: {e}")
        print("Please check the file paths and try again.")
        return None, None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None, None

def preprocess_data(X):
    """
    Preprocesses the dataset by handling non-numeric values and scaling numeric columns.
    """
    try:
        # Handle non-numeric columns
        for col in X.columns:
            if X[col].dtype == 'object':
                print(f"Encoding non-numeric column: {col}")
                X[col] = LabelEncoder().fit_transform(X[col].astype(str))

        # Handle missing values by filling with median
        X = X.fillna(X.median())

        return X

    except Exception as e:
        print(f"Error during preprocessing: {e}")
        return None

# Step 2: Feature Extraction via MLP
def extract_features_with_mlp(X_train):
    """
    Extracts deep features using a Multi-Layer Perceptron (MLP).
    """
    # Scaling the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_train)

    # Define the MLP model
    model = Sequential([
        Dense(64, activation='relu', input_dim=X_scaled.shape[1]),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dense(8, activation='relu'),
        Dense(X_scaled.shape[1], activation='linear')  # Output layer for feature extraction
    ])

    model.compile(optimizer='adam', loss='mse')
    model.fit(X_scaled, X_scaled, epochs=50, batch_size=32, verbose=1)

    # Extract deep features
    deep_features = model.predict(X_scaled)
    return deep_features

# Step 3: Classification using XGBoost
def train_and_classify_with_xgboost(X_train, y_train, X_test):
    """
    Trains an XGBoost classifier and predicts on the test dataset.
    """
    # Train XGBoost classifier
    xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
    xgb.fit(X_train, y_train)

    # Make predictions
    y_pred = xgb.predict(X_test)
    y_pred_prob = xgb.predict_proba(X_test)[:, 1]  # Probabilities for ROC curve
    return xgb, y_pred, y_pred_prob

# Step 4: Plot Learning Curves
def plot_learning_curves(estimator, X_train, y_train):
    """
    Plots learning curves for the given estimator.
    """
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X_train, y_train, cv=5, scoring='accuracy',
        train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1
    )

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.figure()
    plt.title("Learning Curve")
    plt.xlabel("Training Examples")
    plt.ylabel("Accuracy Score")
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training Score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-Validation Score")
    plt.legend(loc="best")
    plt.show()

# Step 5: Plot ROC Curve
def plot_roc_curve(y_test, y_pred_prob):
    """
    Plots the ROC curve and calculates the AUC.
    """
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f"ROC curve (AUC = {roc_auc:.2f})")
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Receiver Operating Characteristic (ROC) Curve")
    plt.legend(loc="lower right")
    plt.show()

# Step 6: Model Evaluation & Visualization
def evaluate_model(classifier, X_test, y_test, y_pred, y_pred_prob):
    """
    Evaluates the model using confusion matrix, classification report, accuracy score, and visualizations.
    """
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(cm)

    # Classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # Accuracy score
    print("Accuracy Score:", accuracy_score(y_test, y_pred))

    # Plot confusion matrix
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Fake', 'Genuine'], yticklabels=['Fake', 'Genuine'])
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.title('Confusion Matrix Heatmap')
    plt.show()

    # Plot ROC curve
    plot_roc_curve(y_test, y_pred_prob)

# Main Execution: End-to-End Pipeline
if __name__ == "__main__":
    # Step 1: Read and preprocess data
    real_users_path = "/content/users.csv"
    fake_users_path = "/content/fusers.csv"
    X, y = read_datasets(real_users_path, fake_users_path)

    # Ensure datasets were loaded successfully
    if X is not None and y is not None:
        # Preprocess the dataset
        X = preprocess_data(X)

        # Split data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        # Step 2: Feature extraction via MLP
        deep_features_train = extract_features_with_mlp(X_train)
        deep_features_test = extract_features_with_mlp(X_test)

        # Combine deep features with original features
        X_train_combined = np.hstack((deep_features_train, X_train))
        X_test_combined = np.hstack((deep_features_test, X_test))

        # Step 3: Classification using XGBoost
        classifier, y_pred, y_pred_prob = train_and_classify_with_xgboost(X_train_combined, y_train, X_test_combined)

        # Step 4: Plot Learning Curves
        plot_learning_curves(classifier, X_train_combined, y_train)

        # Step 5: Model evaluation and visualization
        evaluate_model(classifier, X_test_combined, y_test, y_pred, y_pred_prob)
```
